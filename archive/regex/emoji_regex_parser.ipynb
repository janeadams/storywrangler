```python
def get_svg(q): #give query in format "\\uXXXX"
    try:
        query = q.split("\\u")[1].lower()
        url = str("https://www.fileformat.info/info/unicode/char/"+query+"/index.htm")
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
        svg_url_snip = soup.find(text="Outline (as SVG file)").parent['href']
        svg_url = str("https://www.fileformat.info"+svg_url_snip)
    except:
        print("Couldn't find url for ",q)
        svg_url = "nourl"
    return svg_url
```

```python
def get_name(q):
    url = get_svg(q)
    try:
        url_snip = url.split("/")[-1]
        name = url_snip.split(".")[0]
        name = str(":"+name+":")
        #regex = "/(...*)\.svg"
        #name = re.search(regex, url_snip)
    except:
        print("Couldn't find name for ",q)
        name = "noname"
    return url, name
```
